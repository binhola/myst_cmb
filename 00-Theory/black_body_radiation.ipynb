Black body radiation is classified as a thermal radiation, the radiation emitted by matter in **thermal equilibrium**. 

Such radiation can be obtained by keeping an isolated enclosure at temperature $T$ until equilibrium has been reached. A small hole, then, can be opened to measure the radiation inside without disturbing equilibrium. Using the fact that *photons are massless* and some *thermodynamic arguments*, some important properties of blackbody radiation can be derived.

*Photons are massless*: they can be created and destroyed in arbitrary number by the walls of the container. For practical reasons, self-interactions between photons are neglected. There are no conservation of photon number, the number will be determined by equilibrium at temperature $T$.

Defining the specific intensity $I_\nu$ as the Planck function $B_\nu(T)$:

```{math}
I_\nu(T) \equiv B_\nu(T)
```

```{tip} Specific intensity dependence on temperature
:open:
Specific intensity $I_\nu$ only depends on temperature. This fact can be proved by joining two containers of the same temperature of arbitrary geometries and placing a filter between the two, which passes only a single frequency $\nu$. If $I_\nu \neq I_\nu'$, energy will flow spontaneously between two enclosures leading to the violation the second law of thermodynamics.
```

## Kirchhoff's Law for Thermal Emission

Consider an element of some thermally emitting material at temperature $T$, depending only on its temperature and internal properties. Place this at the opening of a blackbody enclosure, the source function of the material is $S_\nu$. The new configuration is also a blackbody enclosure at $T$. Thus, we have:
```{math}
S_\nu = B_\nu(T)
```
*Kirchhoff's law* is written as
```{math}
j_\nu = \alpha_\nu B_\nu(T)
```
where $j_\nu$ is the emission coefficient of the material.

```{note} Radiative transfer equation
:open:
The effects of emission and absorption on the specific intensity along a ray are described as:

\begin{equation}
\dfrac{dI_\nu}{ds} = -\alpha_\nu I_\nu + j_\nu
\end{equation}
where
- $\alpha_\nu$ is the absorption coefficient
- $j_\nu$ is the emission coefficient

**Emission only:** $\alpha_\nu = 0$. 
$$
\dfrac{dI_\nu}{ds} = j_\nu
$$
which has solution
$$
I_\nu(s) = I_\nu(s_0) + \int_{s_0}^s j_\nu(s') ds'
$$
The increase in brightness is equal to the emission coefficient integrated along the line of sight.

**Absorption only:** $j_\nu = 0$. 
$$
\dfrac{dI_\nu}{ds} = -\alpha_\nu I_\nu
$$
which has solution
$$
I_\nu(s) = I_\nu(s_0) e^{-\int_{s_0}^s \alpha_\nu(s')ds'}
$$
The brightness decreases along the ray by the exponential of the absorption coefficient integrated along the line of sight.
```

```{note} Optical depth
:open:
Define *optical depth* using the absorption coefficient $\alpha_\nu$:
$$
d\tau_\nu = \alpha_\nu ds
$$
or
$$
\tau_\nu = \int_{s_0}^s \alpha_\nu(s') ds'
$$
- **Optical thick** (opaque): $\tau_\nu > 1$
- **Optical thin** (transparent): $\tau_\nu < 1$
```

```{note} Source function
:open:
The radiative transfer can be written as
$$
\dfrac{dI_\nu}{d\tau_\nu} = - I_\nu + S_\nu
$$
where the *source function* $S_\nu$ is defined as
$$
S_\nu \equiv \dfrac{j_\nu}{\alpha_\nu}
$$
```

```{hint} Optical depth and source function
:open:
The source function $S_\nu$ is often a simpler physical quantity than emission coefficient. Also, the optical depth reveals more clearly the important intervals along a ray as far as radiation is concerned. Therefore, $\tau_\nu$ and $S_\nu$ are often used instead of $\alpha_\nu$ and $j_\nu$.
```

The transfer equation for thermal radiation is
```{math}
\dfrac{dI_\nu}{d\tau_\nu} = -I_\nu + B_\nu(T)
```
- **Thermal radiation:** $S_\nu = B_\nu$
- **Blackbody radiation:** $I_\nu = B_\nu$

Thermal radiation becomes blackbody radiation only for optically thich media.

## Thermodynamics of Blackbody radiation

The first law of thermodynamics:
```{math}
dQ = dU + PdV
```
where $Q$ is heat and $U$ is total energy.
The second law of thermodynamics:
```{math}
dS = \dfrac{dQ}{T}
```
where $S$ is entropy. 

We can obtain *Stenfan Boltzmann* law for energy density $u$ and flux $F$ as
```{math}
u(T) = a T^4 \quad \text{or} \quad F = \sigma T^4
```
where
- $\sigma \equiv \dfrac{ac}{4} = 5.67 \times 10-5 \, \mathrm{erg \, cm^{-2} \, deg^{-4} \, s^{-1}}$
- $a = 7.56 \times 10^{-15} \, \mathrm{erg \, cm^{-3} \, deg^{-4}}$

```{note} Stefan Boltzmann law (XXX Draft)
:open:
We have $U=uV$ and $p = \dfrac{u}{3}$

$$
dS = \dfrac{V}{T} \dfrac{du}{dT} dT + \dfrac{4u}{3T}dV
$$

Since $dS$ is a perfect differential
$$
\left( \dfrac{\partial S}{\partial T}\right)_{V} = \dfrac{V}{T} \dfrac{du}{dT} \quad \quad \left( \dfrac{\partial S}{\partial V}\right)_{T} = \dfrac{4}{3} \dfrac{u}{T}
$$
We use $\dfrac{\partial^2 S}{\partial T \partial V} = \dfrac{\partial^2 S}{\partial V \partial T}$ to get
$$
\log u = 4 \log T + \log a
$$
```

We, then, have the adiabatic expansion for blackbody radiation as
```{math}
TV^{1/3} = \text{const} \quad \quad PV^{4/3} = \text{const}
```

## Planck spectrum
Consider a photon of frequency $\nu$ propagating in direction $\mathbb{n}$ inside a box. The wave vector of the photon is 
$$
\mathbb{k} = \dfrac{2\pi}{\lambda} \mathbb{n} = \dfrac{2\pi \nu}{c} \mathbb{n}
$$

For each dimension of box $L_x$, $L_y$ and $L_z$ is much longer than a wavelength, the photon can be represented by standing waves in the box.
