```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
import scienceplots

plt.style.use(["science"])
```

```python
H0 = 70 # km/s/Mpc
z_rec = 1100
z_eq = 3400

O_m = 0.3
O_r = O_m / (1+z_eq)
O_de = 1 - O_m

def H(z):
    return H0 * np.sqrt(O_m * (1 + z)**3 + O_r * (1 + z)**4 + O_de)

def I(z1, z2):
    integral, _ = quad(lambda z: 1/H(z), z1, z2)
    return integral
```

```python
#| label: I_evo
z_vals = np.linspace(0, 10000, 1000)
I_vals = [I(0, z) for z in z_vals]

plt.figure(figsize=(6,4))
plt.plot(z_vals, I_vals, label=r"$I(0,z)$")
plt.axvline(z_rec, color = 'red', label="$z_{rec}$ = 1100")
plt.xlabel("z")
plt.ylabel(r"$I(0, z) = \tau - \tau_i$")
plt.legend()
plt.xscale("log")
plt.yscale("log")
plt.xlim(10, 10000)
# plt.grid(True)
plt.tight_layout()
# plt.savefig("I_of_z.png") # save figure
plt.show()
```
